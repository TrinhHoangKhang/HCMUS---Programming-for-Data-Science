## Exercises 1

Use `re.search` to find whether a string contains a phone number. The pattern that you write should detect a phone number in the following strings.  
```
"Call me at 382-384-3840."  
"my number is (510) 849-3519. Call me!"
```  
And not find a match in the following strings. 
```
"my number is 510-849-35192"  
"here’s my number: 510-849.3519"
``` 
Consider making your own tests as well  

````python
import re
````

````python
string1 = "Call me at 382-384-3840."
string2 = "my number is (510) 849-3519. Call me!"
string3 = "my number is 510-849-35192" 
string4 = "here's my number: 510-849.3519"
````

````python
question_1_strings = [string1, string2, string3, string4]

pattern = re.compile(r'\(?\d{3}\)?[-|\s]\d{3}-\d{4}\b')
for string in question_1_strings:
    result = re.search(pattern, string)
    if result:
        print(result.group())
    else:
        print("Not found")
````

## Exercise 2

Use `re.sub` to alter the string below so that the dates have a common format that uses a dash for the day, month, and year separator.  
```
03/12/2018, 03.13.18, 03/14/2018, 03:15:2018
```

````python
# YOUR CODE HERE
date_list = ["03/12/2018", "03.13.18", "03/14/2018", "03:15:2018"]
pattern = re.compile(r'[/.:]')
replace = r'-'
for date in date_list:
    new_date = re.sub(pattern, replace, date)
    print(new_date)
````

````python
# Extract name and email address of sender
# YOUR CODE HERE
names = []
emails = []

pattern = re.compile(r'^From:\s+"(.+)"\s+<(.+)>', re.MULTILINE)
results = re.finditer(pattern, fh)
for result in results:
    name = result.group(1)
    email = result.group(2)
    names.append(name)
    emails.append(email)

# In kết quả
print("Names: ")
for name in names:
    print(name)

print("Emails: ")
for email in emails:
    print(email)
````
